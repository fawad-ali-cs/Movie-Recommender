### Saving movie titles

```python
movie_df = pd.read_csv('ml-latest-small/movies.csv')

movies_rated = ratings['movieId'].unique()
movie_df = movie_df[movie_df['movieId'].isin(movies_rated)].reset_index(drop=True)
```

```python
titles = list(movie_df['title'])
```

```python
len(titles)
```

```python
titles
```

```python
import json

with open("data/titles.json", "w") as outfile:
    json.dump(titles, outfile)
```
