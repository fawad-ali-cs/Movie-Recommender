```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
precision_results = {
    "GBPR" : 0.14541,
    "LREC" : 0.25016,
    "TSUISIMCF" : 0.23902,
    "TSUISIMCF \n and \n TF-IDF" : 0.26951,
    "LREC \n and \n TF-IDF" : 0.25290
}

recall_results = {
    "GBPR" : 0.04791,
    "LREC" : 0.12363,
    "TSUISIMCF" : 0.03937,
    "TSUISIMCF \n and \n TF-IDF" : 0.04440,
    "LREC \n and \n TF-IDF" : 0.12498
}

F1_results = {
    "GPPR" : 0.07207,
    "LREC" : 0.16548,
    "TSUISIMCF" : 0.06761,
    "TSUISIMCF \n and \n TF-IDF" : 0.076235,
    "LREC \n and \n TF-IDF" : 0.16729
}
```

```python
precision_results = pd.Series(precision_results)
precision_results
```

```python
recall_results = pd.Series(recall_results)
recall_results
```

```python
F1_results = pd.Series(F1_results)
F1_results
```

```python
plt.figure(figsize = (8, 7))
plt.bar(precision_results.index, precision_results, color="darkred")

plt.title("Best Precision for Each Model", size=14)
plt.xlabel("Recommender System", size=13)
plt.ylabel("Precision", size=13)

plt.show()
```

```python
plt.figure(figsize = (8, 7))
plt.bar(recall_results.index, recall_results, color="teal")

plt.title("Best Recall for Each Model", size=14)
plt.xlabel("Recommender System", size=13)
plt.ylabel("Recall", size=13)

plt.show()
```

```python
plt.figure(figsize = (8, 7))
plt.bar(F1_results.index, F1_results, color="darkgreen")

plt.title("Best F1-Score for Each Model", size=14)
plt.xlabel("Recommender System", size=13)
plt.ylabel("F1-Score", size=13)

plt.show()
```
